```python
import yfinance as yf

# Define the ticker symbol for Brent Crude Oil
ticker_symbol = "BZ=F"

# Fetch data
data = yf.download(ticker_symbol, start="2010-01-01", end="2024-12-31")

# Display the first few rows
print(data.head())
```

```python
import matplotlib.pyplot as plt

# Statistical Summary
print(data.describe())

# Time Series Plot
plt.figure(figsize=(14, 7))
plt.plot(data['Close'])
plt.title('Brent Crude Oil Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.grid(True)
plt.show()

# Histogram of the closing prices
plt.figure(figsize=(10, 5))
plt.hist(data['Close'], bins=50, alpha=0.75)
plt.title('Distribution of Brent Crude Oil Closing Prices')
plt.xlabel('Closing Price')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Calculating daily returns as a proxy for volatility
data['Daily_Return'] = data['Close'].pct_change()
data['Volatility'] = data['Daily_Return'].rolling(window=30).std() * (252**0.5)  # Annualize the standard deviation

# Plotting volatility
plt.figure(figsize=(14, 7))
plt.plot(data['Volatility'])
plt.title('Volatility of Brent Crude Oil Prices')
plt.xlabel('Date')
plt.ylabel('Volatility')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import yfinance as yf

# Fetch data
ticker_symbol = "BZ=F"
data = yf.download(ticker_symbol, start="2010-01-01", end="2020-12-31")

# 1. Statistical Summary
print("Statistical Summary:\n", data['Close'].describe())

# 2. Time Series Plot of Closing Prices
plt.figure(figsize=(14, 7))
plt.plot(data['Close'], label='Closing Price')
plt.title('Brent Crude Oil Closing Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.show()

# 3. Histogram of the Closing Prices
plt.figure(figsize=(10, 5))
plt.hist(data['Close'], bins=50, alpha=0.75, color='blue')
plt.title('Distribution of Brent Crude Oil Closing Prices')
plt.xlabel('Price (USD)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# 4. Volatility Calculation and Plot
data['Daily_Return'] = data['Close'].pct_change()
data['Volatility'] = data['Daily_Return'].rolling(window=30).std() * (252**0.5)  # Annualizing the standard deviation

plt.figure(figsize=(14, 7))
plt.plot(data['Volatility'], label='Volatility')
plt.title('Volatility in Brent Crude Oil Prices')
plt.xlabel('Date')
plt.ylabel('Volatility')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import yfinance as yf

# Fetch data
ticker_symbol = "BZ=F"
data = yf.download(ticker_symbol, start="2010-01-01", end="2024-12-31")

# Calculate moving averages
data['MA50'] = data['Close'].rolling(window=50).mean()  # 50-day moving average
data['MA200'] = data['Close'].rolling(window=200).mean()  # 200-day moving average

# Plotting the prices with moving averages
plt.figure(figsize=(14, 7))
plt.plot(data['Close'], label='Closing Price')
plt.plot(data['MA50'], label='50-Day MA')
plt.plot(data['MA200'], label='200-Day MA')
plt.title('Brent Crude Oil Prices with Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import yfinance as yf
import seaborn as sns
import pandas as pd

# Fetch data
ticker_symbol = "BZ=F"
data = yf.download(ticker_symbol, start="2010-01-01", end="2024-12-31")

# Extract year and month from the index
data['Year'] = data.index.year
data['Month'] = data.index.month

# Group by month to see average monthly prices
monthly_prices = data.groupby('Month')['Close'].mean()

# Plot
plt.figure(figsize=(10, 5))
sns.barplot(x=monthly_prices.index, y=monthly_prices.values, palette="viridis")
plt.title('Average Monthly Closing Prices for Brent Crude')
plt.xlabel('Month')
plt.ylabel('Average Closing Price (USD)')
plt.xticks(ticks=range(0, 12), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.show()
```

```python
import yfinance as yf
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt

# Fetch data
ticker_symbol = "BZ=F"
data = yf.download(ticker_symbol, start="2010-01-01", end="2020-12-31")

# Prepare data
data = data['Close'].resample('MS').mean()  # Monthly mean resampling

# Fit the ARIMA model
model = ARIMA(data, order=(5,1,0))  # ARIMA(p,d,q) where p is the order of the autoregressive part, d is the degree of differencing, and q is the order of the moving average term.
model_fit = model.fit()

# Forecast
forecast = model_fit.forecast(steps=12)  # Forecast the next 12 months

# Plot the results
plt.figure(figsize=(14, 7))
plt.plot(data, label='Historical Monthly Mean Close Price')
plt.plot(forecast, label='Forecasted Price', color='red')
plt.title('ARIMA Forecast of Monthly Mean Closing Price')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.show()
```
